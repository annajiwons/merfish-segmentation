```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
csv_names = ['pretrained_add5', 'pretrained_add10', 'stardist', 'aug_fliprot', 
             'aug_intensity', 'aug_noise', 'smfish', '4t1_stardist']


fig, ax = plt.subplots(2, 4, figsize = (6 * 4, 6 * 2))

for i in range(len(csv_names)):
    df = pd.read_csv(f'loss_epoch/{csv_names[i]}.csv')
    df = df.drop(['Wall time'], axis = 1)
    sns.lineplot(ax = ax[i // 4][i % 4], x = 'Step', y = 'Value', color = 'black', data = df)
    ax[i // 4][i % 4].set_title(f'Model {i + 2}')
    ax[i // 4][i % 4].set(xlabel = 'Epoch', ylabel = 'Loss')
    sns.despine()

plt.show()
```
